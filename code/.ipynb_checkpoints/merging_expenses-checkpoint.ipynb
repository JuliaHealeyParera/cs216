```python
import pandas as pd 
import numpy as np
import math
from fuzzywuzzy import process
```

```python
expenses = pd.read_csv("../data/Sport_Data_2003_2004_2005_2006_2007_2008_2009_2010_2011_2012_2013_2014_2015_2016_2017_2018_2019_2020_2021_2022.csv")
teamsa_total = pd.read_csv("teamsa_total.csv")
teamsb_total = pd.read_csv("teamsb_total.csv")

expenses.columns.values[40] = "expenses_women"
expenses.columns.values[39] = "expenses_men"
expenses.columns.values[0] = "survey_year"
expenses.columns.values[3] = "institution_name"
col_names = ['expenses_women', 'expenses_men', 'survey_year', 'institution_name']
expenses['institution_name'] = expenses['institution_name'].apply(lambda x : x.replace("saint", 'st'))
expenses['institution_name'] = expenses['institution_name'].apply(lambda x : x.replace("Alabama State", 'Alabama'))
expenses['institution_name'] = expenses['institution_name'].apply(lambda x : x.replace("alabama state", 'alabama'))
expenses['institution_name'].unique()
```

```python
series1 = pd.Series(teamsa_total['TEAM']).astype(str)
series2 = pd.Series(expenses['institution_name'].unique()).astype(str)

def find_closest_match(x, threshold=80):
    match = process.extractOne(x, series2)
    return match[0] if match and match[1] >= threshold else None

closest_matches = series1.apply(find_closest_match)

win_percent_dict = expenses.set_index('Team').to_dict(orient='index')

def map_to_win_percent(match):
    if match is None:
        return {}  
    return win_percent_dict.get(match, {}) 

mapped_data = closest_matches.apply(map_to_win_percent)

mapped_df = pd.DataFrame(list(mapped_data), index=closest_matches.index)

#teamsa_total['Closest Match'] = closest_matches
#teamsa_total = pd.concat([teamsa_total, mapped_df], axis=1)

#with_expenses = pd.merge(teamsA, win_percent, how='left', left_on='Closest Match', right_on='Team')
```
