```python
import pandas as pd 
import numpy as np
import math
from fuzzywuzzy import process
```

```python
teamsA = pd.read_csv("../data/Tournament Matchups.csv")
teamsB = pd.read_csv("../data/Tournament Matchups.csv")
win_percent = pd.read_csv("../data/merged_win_loss_stats.csv")
```

```python
series1 = pd.Series(teamsA['TEAM']).astype(str)
series2 = pd.Series(win_percent['Team']).astype(str)

def find_closest_match(x, threshold=80):
    match = process.extractOne(x, series2)
    return match[0] if match and match[1] >= threshold else None

closest_matches = series1.apply(find_closest_match)

win_percent_dict = win_percent.set_index('Team').to_dict(orient='index')

def map_to_win_percent(match):
    if match is None:
        return {}  
    return win_percent_dict.get(match, {}) 

mapped_data = closest_matches.apply(map_to_win_percent)

mapped_df = pd.DataFrame(list(mapped_data), index=closest_matches.index)

teamsA['Closest Match'] = closest_matches
teamsA = pd.concat([teamsA, mapped_df], axis=1)

teamsa_total = pd.merge(teamsA, win_percent, how='left', left_on='Closest Match', right_on='Team')
```

```python
series3 = pd.Series(teamsB['TEAM']).astype(str)
series4 = pd.Series(win_percent['Team']).astype(str)

def find_closest_match(x, threshold=80):
    match = process.extractOne(x, series4)
    return match[0] if match and match[1] >= threshold else None

closest_matches_b = series3.apply(find_closest_match)

win_percent_dict = win_percent.set_index('Team').to_dict(orient='index')

def map_to_win_percent(match):
    if match is None:
        return {}  
    return win_percent_dict.get(match, {})  

mapped_data_b = closest_matches_b.apply(map_to_win_percent)

mapped_df_b = pd.DataFrame(list(mapped_data_b), index=closest_matches_b.index)

teamsB['Closest Match'] = closest_matches_b
teamsB = pd.concat([teamsB, mapped_df_b], axis=1)

teamsb_total = pd.merge(teamsB, win_percent, how='left', left_on='Closest Match', right_on='Team')

print(teamsb_total)
```
